```python
from functools import reduce

with open("file.txt", "r") as file:
    
    # Counting number of lines
    countLines = file.readlines()
    print(countLines)
    l = len(countLines)
    print(f"Total number of lines: {l}\n")

    # Counting number of words
    countWords = list(map(lambda x : len(x.split()), countLines))
    print(countWords)
    countWords = reduce(lambda x, y : x + y, countWords)
    print(f"Total number of words: {countWords}\n")

    # Counting number of characters
    countChars = list(map(str.strip, countLines))
    print(countChars)
    countChars = list(map(lambda x : x.replace(" ", ""), countChars))
    print(countChars)
    countChars = list(map(lambda x : len(x), countChars))
    print(countChars)
    countChars = reduce(lambda x, y : x + y, countChars)
    print(f"Total number of characters {countChars}")

    # saving all the counts in file.txt
    with open("file.txt", "a") as file:
        file.write(f"Total Number of lines: {l}\n")
        file.write(f"Total number of words: {countWords}\n")
        file.write(f"Total number of characters: {countChars}\n")
```

## Methods

```python
# base / parent class
class Phone:
    # class attribute
    category = "electronics"

    # constructor
    def __init__(self, model, battery, camera, battery_percentage = 100):
        # instance attribues
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    # instance methods can access both class and instance attribute
    # instance method
    def photoCapture(self, capture):
        if self.battery_percentage < 0:
            print("No charge")
        self.battery_percentage -= capture
        print(f"Phote captured in {self.model}")

    # instance method
    def charge(self, hour):
        print(f"charge duration {hour} hour(s)")

    @classmethod
    def info(self):
        # class methods can only access class attribues
        print(self.category)

apple = Phone("apple", 3000, 30)
samsung = Phone("S7", 40000, 40)
apple.charge(2)
apple.photoCapture(10)
print(f"apple battery percentage - {apple.battery_percentage}")
print(f"samsung battery percentage - {samsung.battery_percentage}")
apple.info()
```

## Inheritence

```python
# derived / child class
class SmartPhone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor

    def charge(self, hours):
        print("fast charging in progress")
        super().charge(hours)

nothing = SmartPhone("nothing", 1000, 10, "SnapDragon")
print(f"model - {nothing.model}")
print(f"battery - {nothing.battery}")
print(f"battery - {nothing.battery}")
print(f"camera - {nothing.camera}")
print(f"processor - {nothing.processor}")
print(f"battery percentage - {nothing.battery_percentage}")

nothing.photoCapture(30)
print(f"battery percentage after photo capture - {nothing.battery_percentage}")

nothing.charge(10)
```

### Multiple Inheritence

```python
class CoolingMechanism:
    def __init__(self, method):
        self.method = method

    def coolingOn(self):
        print(f"The system is being cooled by {self.method}")
```

```python
class SmartPhone_CoolingMechanism(SmartPhone, CoolingMechanism):
    def __init__(self, model, battery, camera, processor, method):
        SmartPhone.__init__(self, model, battery, camera, processor)
        CoolingMechanism.__init__(self, method)

xiaomi = SmartPhone_CoolingMechanism("xiaomi", 1000, 10, "SnapDragon", "Nitro")
print(f"model - {xiaomi.model}")
print(f"battery - {xiaomi.model}")
print(f"camera - {xiaomi.camera}")
print(f"processor - {xiaomi.processor}")
print(f"battery - {xiaomi.battery_percentage}")
xiaomi.photoCapture(10)
print(f"battery after photo capture - {xiaomi.battery_percentage}")
xiaomi.charge(10)
xiaomi.coolingOn()
```

## Polymorphism

```python
class Camera:
    def __init__(self, name):
        self.name = name

    def capture(self):
        print("a photo is captured")

class Smart(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # method overriding
    def capture(self):
        print(f"A photo is being captured by a {self.name}")

class DSLR(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # method overriding
    def capture(self):
        print(f"A photo is being captured by a {self.name}")

class Drone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # method overriding
    def capture(self):
        print(f"A photo is being captured by a {self.name}")


smart = Smart("smart_phone", 10)
dslr = DSLR("DSLR", 10)
drone = Drone("Drone", 10)

smart.capture()
dslr.capture()
drone.capture()
```

## Encapsulation

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei    # private

    def charge(self):
        print("Phone is charging")

iphone = Mobile("iphone", 17, 102030)

print(f"Default IMEI: {iphone.__imei}")
iphone.imei = "1x1x1x"
print(f"Updated IMEI: {iphone.__imei}")
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.name = name
        self.model = model
        self.__imei = imei    # private

    def getter(self):
        return self.__imei

iphone = Mobile("iphone", 17, 102030)
print(f"Default IMEI: {iphone.getter()}")
iphone.__imei = "1x2x3x"
print(f"Changed IMEI: {iphone.getter()}")
# imei didn't change and now can only be accessed by getter method
```

```python
class Mobile:
    def __init__(self, name, model, imei):
        self.__imei = imei    # private

    def getter(self):
        return self.__imei
    
    def setter(self, imei):
        self.__imei = imei

iphone = Mobile("iphone", 17, 102030)
print(f"Default IMEI: {iphone.getter()}")
iphone.setter("1x2x3x")
print(f"Changed IMEI: {iphone.getter()}")
```
