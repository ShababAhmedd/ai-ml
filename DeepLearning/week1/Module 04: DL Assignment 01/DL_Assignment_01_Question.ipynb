# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write** Answer 01:


```python
'''
Perceptron is the simplest form of Artificial Neural Network (ANN) which is used for binary classification. Thus, it outputs either 0 or 1 based
on multiple input values.

Inputs and weights:
There can be multiple inputs namely features and along with each feature corresponding weights are associated with it. Weight dictates the importance
of each feature. Input and their corresponding weights are multiplied before they are passed on to the activation function. Z = x_1*W_1 + ... + X_n*W_n.

Bias:
Before finally it enters activation function, bias b is added to this weighted sum. Bias helps to adjust the decision boundary.
Z = x_1*W_1 + ... + X_n*W_n + b

Activation Function:
After it is passed on to the activation function, it convertes values into values of 0/1 i.e. binary classification output.
Step function ==> 1 if z>= 0, else 0
'''
```

# Question 03: [ Marks 15 ]

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:

```python
'''
Because perceptron seperates classes using a linear straight line. As far as linear seperability is concerned, a data set can only be classified using 
perception if and only if the data points within that data set can be seperated using a straight line. Perceptron exactly does that by creating a 
decision boundary. 

E.g.
---AND Gate---
Here, output is only 1 when both inputs are 1. So, using a striaght line this one can be seperated which is situated right across the origin
if the points are plotted in a graph

--OR Gate--
Here, output is 1 iff any one of the input is 1. If plotted in a graph, three 1s are on the positive side of the decision boundary. One 0 is on the negative side.
Again, can be seperated using linear striaght line.

--XOR--
This, can not be seperated since no matter which way line is drawn there will atleast one point of other class that gets into the positive or negative
side of the decision boundary. For this a striaght line wont work. For this, we need some model that can form a non-linear decision boundary. 

To sum up, perceptron can only classify problems which are linearly seperable since perceptron itself forms a striaght line decision boundary. 
For problems like XOR, non-linear classifiers are needed. 
'''
```

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:

```python
'''
---Perception Weight Update Formula/Rule---
w_new = w_old + eta*(y - y_hat)*x_i (b is also updated in similar manner)
where, 
w = wieght of input x_i
eta = learning rate
y = actual class label
y_hat = predicted class label
x_i = input feature

eta dictates how fast or slow the update will take place.
error aka y - y_hat dictates whether the update will increase or decrease the weight
x_i refers to the input ith feature whose corresponding weight will either increase or decrease from its previous value.
In short, this helps to update the weights in such a manner that the decision boundary will then correctly align itself so that it can 
correctly classify the classes. 

If the existing weight gives correct prediction then no weight updatation is required. Since the existing weights correctly classify the classes. 
Thus, no change is required. 
On the other hand, if the prediction is incorrect, then weight updation takes place so that the perception can classify correctly by correctly aligning 
its decision boundary. 
'''
```
