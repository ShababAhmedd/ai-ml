# Question 01: [ Marks 10 ]
Explain gradient descent in detail. What is the main idea behind it? How does it help in minimizing the loss function? Explain the role of learning rate and describe what happens if the learning rate is too high or too low.

**Write** Answer 01:


```python
'''
Gradient Descent is an optimizing algorithm which is used to find the minimum value of a cost function. It works iteratively.
The minimum value is found by adjusting the learnable parameteres i.e. weights and biases. Initially, the values of learnable
are set random which gradually gets adjusted during training the model. With respect to the learnable parameters, the loss is calculated 
and a lower loss is found by adjusting the values of parameters which gives the the minimum loss. Thus, the parameters are updated 
in a manner that gives a lower loss. Since this is a iterative algorithm, the process is run n number of epochs or until the loss 
becomes minimized. 
m_new = m_old - eta * del L/del m
b_new = b_old - eta * del L/del b

Here, the derivative of the function L(m, b) = 1/n sum(i=1 to n) [y-(mx_i+b)]^2 with respect to m or b gives us the rate of change. It gives us the direction
of the steepest increase and if we move in the opposite direction of the slope we move towards a lower loss. So, we use slope to help us reach to minima. 

Let us first start by defining the term loss function. Loss function is something that measures the error of a model's prediction. So, what this algo does
is it calculates the derivate of the loss function as given above w.r.t to learnable parameters. The learnable parameters are then
adjucted in a manner that reduces the overall loss. As mentioned above, since this is a iterative algorithm, the process is repeated
for the given number of epochs or when until the error becomes very small. As a result, with every iteraiton, not only the loss decreases
but the prediction becomes more accurate. At the sufficient number of the epoch, the model ideally converges and settles at minima. 

Learning rate controls the size of the update for each of the learnable parameters in each epoch. 
A high learning rate means the update will be large. In such cases, the model may fail to converge and it may fail to reach the minima. 
The loss might increase. 
A low learning rate means that the update will be very small in each epoch. The training will take too long since convergence will be very slow. 
'''
```

Explain the sigmoid activation function. Where is it used? What are its advantages and disadvantages? What problem does it cause during training?

## Write Answer 04:

```python
'''
The sigmoid function is defined as sigmoid(x) = 1/(1 + e^-x). The ouput of this function is between 0 and 1 no matter what the input is.
The function is also differentiable at every point. When x is 0, y is 0.5. As a result, 0.5 commonly choosen as a threshold value. 

Generally speaking, sigmoid function is used for binary classification in the last layer which gives the probability for each binary classes. 
Machine learning models like Logistic Regression uses sigmoid activation function.  

Advantages:
Can be interpreated as probability - bounded by 0 and 1
Differentiable and smooth.
Order of input is preserved

Disadvantages:
gradients become very small at high or low inputs.
vanishing gradient problem
not zero centered - so gradient updates can fluctuate

The derivative of sigmoid approaches 0 for large positive or negative inputs. 
Further more, gradients become very small during backpropagation.
If the network is very deep, weights updates very slowly.  
'''
```

# Question 05: [ Marks 10 ]
Explain the softmax activation function in simple words. Why is it used in multi-class classification problems? How is it different from sigmoid?

## Write Answer 05:

```python
'''
Softmax activation function is defined as softmax(z_i) = e^z_i/sum(j=1 to k) e^z_j
Say, before applying activation the values of z for different class are as follows: z = [2, 1, 0]
Say, we apply this activation for the first class: softmax(2) = e^2/(e^2 + e^1 + e^0)
So, we can say, what this function does is it converts a vector comprised of logits into output probabilites by taking each of the number from a vector.
p.s. all the probabilites in the denominator (e^2 + e^1 + e^0) which is used to normalize sums up to 1.

This activation function outputs all the probabilites at once where the highest probability indicates the predicted class. Since, it ouputs all the 
probabilites at once, theres room for comparison among all the predicted classes. Additionally, it works well with losses like 
categorical cross entropy. 

Sigmoid is commonly used for binary classification whereas Softmax is used for multi class classification. As sigmoid is for binary classification, 
it outputs only one probability. In contrast, Softmax outputs all the probabilities for each of the respective classes. Each of the outputs
by sigmoid is independent whereas the output of softmax is dependent upon each other. In softmax, probabilites of all the respective classes sum
up to 1 which is unlikely the case for sigmoid where only one output is given and it doesnt necessarily have to be 1. So, an output of sigmoid is independent
whereas softmax provides probability distribution across all the classes. 
'''
```

# Question 06: [ Marks 20 ]

## Write Answer 06:

```python
'''
The structure of an Artificial Neural Network (ANN) comprises of input layer, single or multiple hidden layers and lastly one output layer. 
Each of these layers consists of neurons one stacked upon another. The neurons are connected to their neighbouring neurons (i.e. next layer )
via weights. Biases are associated with each neuron. Each of the neuron performs a simple linear mathematical operation. z = sum(w_i*x_i) + b where, 
w_i = weights, x_i = inputs, b = bias.

Input layers receives the raw inputs known as features which passes the inputs forward. The number of neurons within the input layer depends upon
the number of features the data set have. To give an example, if a dataset has n features, the number of neurons in the input layer will also be n. 

Hidden layer is where the learning of most of the complex pattern takes place. When we generally say that the network is deep we imply the network to
have many hidden layers. A deeper network helps to capture more complex patterns. Here, within the hidden layers, weights, biases, activation functions 
are applied on the data via each neuron. Here, the model learns patterns and captures complex features. 

Output layer gives the final prediction. Here, activation functions are carefully choosen which largely depends upon the task that is being done
with the model. For instance, for binary classification, sigmoid function is commonly choosen whereas for multiclass classification softmax is choosen. 

Weights determine the importance of each of the feature. When a input is passed to a neuron, it multiplies each of the input with weights. Bias is 
a value that is generally assigned. For example, you went to rent a place which depends upon the number of bedrooms the apartment has, the floor it is situated 
in etc. But bofore doing that you might need to submit a registration form which requires a charge. In this context, this registration charge is the bias. 
Number of bedrooms, the floor number where the apartment is situated is are features. Also, bias like weight can shift the decidion boundary. 
z = w_1*x_1 + w_2*x_2 + ... + w_n*x_n + b

Activation Functions helps to introduce non-linearity which helps to learn complex features like XOR data where it is not possible to draw a 
decision boundary by a simple linear line. It also controls or dictates how the neuron output will be. 

So, what happens is a layer contains many neurons. so output from these neurons becomes input to neurons of the next layer. This takes place until
the data reaches the output layers which produces the final prediction. This is known as forward propagation. Lets illustrate this using a simple example,
Suppose we want to know if a sutdent passed or not based on number of hours studied(x_1) and the attendance(x_2). So, in the input layer, there will
be two neurons x_1 and x_2. These inputs will be passed onto the hidden layers. Say we have two neurons in the hidden layer where each neuron computes
z = w_1*x_1 + w_2*x_2 + b and applies any of the activaiton function. when these data gets passed onto the output layer it goes through sigmoid activation
function since we are doing a binary classification. If the output of this activation funciton is 0.75, it means there is 75% of the student being passed. 
Roughly, the flow is like input then weights and biases are applied which then goes to the activation function. This then goes to the next layer and same 
things take place until it reaches the output layer where final prediction is given. 

During training, the weights gets adjucted in a way that learns feature importance and helps to predict a more accurate output. Bias adjucts the decision
boundary providing it with flexibility. Activation function helps to introduce non-linearity and when each of the layers with non-linear activation
functions combine they help to form complex decision boundaries. 
'''
```
