```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Q1
df = pd.read_csv('student_IQdata.csv')
df
```

```python
# Q2
plt.scatter(df['Study_Hour'], df['IQ_Score'])
plt.xlabel("Study Hour")
plt.ylabel("IQ Score")
plt.title("Relationship between Study Hour and IQ Score")
plt.show()
```

```python
# Q3
plt.scatter(df['Shoe_Size'], df['Study_Hour'])
plt.xlabel("Shoe Size")
plt.ylabel("Study Hour")
plt.title("Relationship between Shoe Size and Study Hour")
plt.show()
```

```python
# Q4
plt.hist(df['Shoe_Size'], edgecolor='black', bins=20)
plt.xlabel("Shoe Size")
plt.ylabel("Frequency")
plt.title("Histogram of Shoe Size")
```
