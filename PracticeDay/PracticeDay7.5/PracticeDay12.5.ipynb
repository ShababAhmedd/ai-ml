### Question 6

Create a **DataFrame** of employees with columns `"Name"`, `"Department"`, and `"Salary"`.  

- **Sort** the DataFrame by `"Salary"` in **ascending** order.  
- Then sort it by `"Name"` in **descending** order.


````python
df = pd.DataFrame({
    "Name": ["Mashrur", "Safir", "Shabab", "Ahmed"],
    "Department": ["CS", "EEE", "Mechatronics", "Robotics"],
    "Salary": [1, 2, 1, 2],
})

df
````

````python
df.sort_values(["Salary", "Name"], ascending=[1, 0], inplace=True)
df
````

### Question 7

Create a **DataFrame** of **8 students** with columns `"Name"` and `"Score"`.
