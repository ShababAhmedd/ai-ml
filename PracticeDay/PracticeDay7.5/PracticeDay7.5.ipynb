# Problem 1: Common Friends

```python
a_friends = {"Rahim", "Karim", "Sakib", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}

commonFrnd = a_friends.intersection(b_friends)
print(f"Mutual friends: {commonFrnd}")

aUnique = a_friends - commonFrnd
print(f"Unique to A: {aUnique}")

bUnique = b_friends-commonFrnd
print(f"Unique to b: {bUnique}")

totalUniqueFrnd = len(commonFrnd) + len(aUnique) + len(bUnique)
print(f"Total unique friends: {totalUniqueFrnd}")

```

# Problem 2: Sort Students by Marks

```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]
students.sort(key = lambda x : x[1], reverse=True)
print(students)
```
