### min-max scaling

```python
min = df["Total Marks"].min()
max = df["Total Marks"].max()

df["Scaled Marks"] = df["Total Marks"].apply(lambda x : (x-min)/(max-min))

df
```

```python
def grading(x):
    if x >= 250:
        return "A+"
    elif x >= 200:
        return "A"
    else:
        return "P"
    
df["Grade"] = df["Total Marks"].apply(grading)

df
```

### str.contains()

```python
# only showing cities who have the substring "New" in its name
# case sensitive
df.loc[df["City"].str.contains("new")]
```

```python
# no longer case sensitive as case is set to false
df.loc[df["City"].str.contains("new", case=False)]
```

#### regular expression

```python
# city names that start with the substring Los.
df.loc[df["City"].str.contains(r"^los",case=False)]
```

```python
# city names that ends with the substring rk.
df.loc[df["City"].str.contains(r"rk$",case=False)]
```

```python
# names that start with vowel
df.loc[df["Name"].str.contains(r"^[aeiou]", case=False)]
```

```python
# city names that contains the substring either new or loss
df.loc[df["City"].str.contains(r"new|los", case=False)]
```

## DateTime and time delta

```python
df = pd.read_csv("student_completed.csv")
df
```

### changing the data type of EnrollmentDate series to datetime

```python
df["EnrollmentDate"]
```

```python
df["EnrollmentDate"] = pd.to_datetime(df["EnrollmentDate"])
df["FinishedDate"] = pd.to_datetime(df["FinishedDate"])
```

```python
df["EnrollmentDate"]
```

### extracting year, day, month from EnrollmentDate 

```python
df["Enrolment Day"] = df["EnrollmentDate"].dt.day
df["Enrolment Month"] = df["EnrollmentDate"].dt.month
df["Enrolment Year"] = df["EnrollmentDate"].dt.year

df
```

```python
df["Duration Required"] = df["FinishedDate"] - df["EnrollmentDate"]
df
```
